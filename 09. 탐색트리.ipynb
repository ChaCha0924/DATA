## 09. 탐색트리

### 탐색트리란?
#### 탐색트리는 탐색을 위한 트리 기반의 자료구조이다
- 가장 간단한 순차탐색은 비효율적이기 때문에 이진탐색으로 성능을 개선할 수 있지만, 정렬된 데이터에 대해서만 적용 가능
- 이진탐색은 레코드를 한번 정렬하고 탐색이 빈번하게 필요한 응용에는 효율적이지만, 레코드의 삽입과 삭제 자체가 빈번하게 발생한다면 그때마다 정렬에 대한 부담이 있으므로 효율적이지 않을 수도 있음
- 탐색을 위한 가장 빠른 방법은 해싱을 사용하는 것이지만 메모리를 많이 사용하고, 해시 충돌과 이에 따른 오버플로 문제를 해결해야함
- 탐색에 트리구조를 이용할 수도 있음
- 이진탐색트리(BST): 효율적인 탐색을 위한 이진트리 기반의 자료구조
 - 모든 노드는 유일한 키를 갖는다
 - 왼쪽 서브트리의 키들은 루트의 키보다 작다
 - 오른쪽 서브트리의 키들은 루트의 키보다 크다
 - 왼쪽과 오른쪽 서브트리도 이진탐색트리이다
- 트리의 연산들은 보다 단순하게 설계하기 위해 이진탐색트리에서는 보통 중복을 허용하지 않지만 필요하다면 중복된 키를 허용할 수도 있음
- 이진탐색트리의 조건: 왼쪽 서브트리 노드의 키값 < 루트 노드의 키값 < 오른쪽 서브트리 노드의 키값
- 이진탐색트리의 모든 노드들은 중위 순회로 방문해보면 오른차순으로 노드를 방문하게 되는데, 모든 이지남색트리는 이러한 성질을 가지며, 따라서 어느 정도 정렬된 상태를 유지한다고 볼 수 있음
- 이진탐색트리는 이진트리의 한 종류
 - 공백검사나 순회, 노드의 수 등 기본적인 이진트리 연산은 이진탐색트리에서도 동일하게 적용됨
 - 이진탐색트리는 탐색이 목적이므로 탐색연산이 정의되어야 함
- 리스트나 스택 , 큐 등에서와 달리 이진트리에서는 삽입이나 삭제연산을 전혀 고려하지 않음
 - 트리의 경우 노드를 삽입하거나 삭제하는 위치를 명시하는 것이 간단하지 않기 때문
 - 이진탐색트리에서는 이들 연산을 구체화할 수 있음
- 이진탐색트리에서는 삽입 연산, 삭제연산 및 탐색연산에 집중
 - insert(n) 연산을 통해 새로운 노드 n이 저장되어야 하는 트리상의 위치는 단 하나뿐(삭제도 동일)
 - 어떤 노드가 삭제되면 이진탐색트리의 조건을 반드시 유지하는 형태로 트리가 재구성되어야 함
 - 모든 연산이 이진탐색트리의 조건을 유지하면서 처리되어야 함

### 이진탐색트리의 연산
- 이진탐색트리를 위한 노드의 구조는 기본적으로 일반적인 이진트리와 동일
 - 하나의 차이는 이진탐색트리는 탐색을 위한 자료구조이므로 노드의 데이터는 하나의 엔트리, 즉 (탐색키, 키에 대한 값)의 형태가 되어야 함
- 효율적인 탐색이 가능하기 위해서는 트리가 반드시 주어진 키 값에 대해 이진탐색트리의 조건에 맞게 구성되어 있어야 함

#### 탐색 연산
- 키를 이용한 탐색
 - 탐색은 항상 루트노드에서 시작
 - 루트노드와의 비교 결과는 세가지 중 하나이며, 이 과정은 루트의 서브트리에서도 되풀이되어야 함(key ==, < , > 루트의 키 값)
 - 순환으로 구현하는 것이 자연스러움
 - 탐색 연산은 반복 구조를 이용해서도 구현할 수 있음
 - 검사하는 노드의 숫자는 순환 구조를 이요한 경우와 정확히 동일하지만, 실행의 효율성만을 따진다면 반복 구조가 더 우수함
- 값을 이용한 탐색
 - 이진탐색트리는 키 값을 기준으로 정렬되어 있음
 - 트리의 모든 노드를 하나씩 검사하면 되기 때문에 키가 아닌 다른 필드를 이요한 탐색도 가능함
 - 모든노드를 방문하는 순회의 방법에는 제한이 없음(전위, 중위, 후위, 레벨 순회 등 트리의 모든 노드를 방문할 수 있다면 어떤 순회도 상과없음)
 - 그러나 항상 트리의 모든 노드를 검사해야하기 때문에 탐색의 효율은 떨어짐
 - 엔트리의 키가 아닌 값을 이용한 탐색 함수의 경우 전위순회를 이요하는데, 최악의 경우(찾는 값을 갖는 노드가 트리에 없는 경우) 트리의 모든 노드를 검사해야 하므로 시간 복잡도는 O(n)
- 최대와 최소 노드 탐색(반복구조, 순환구조)
 - 이진탐색트리에서는 최대와 최소 키를 가진 노드를 찾는 연산도 가능
 - 최대 키는 트리의 가장 오른쪽 노드에 있고, 최소 키는 트리의 가장 왼쪽 노드에 있음

#### 삽입 연산
- 삽입 연산을 위해서는 먼저 삽입할 노드의 키를 이용한 탐색 과정을 수행해야 하는데, 탐색에 실패한 위치가 바로 새로운 노드를 삽입해야 하는 위치이기 때문
- 탐색이 성공하면 중복된 키를 가진 노드가 이미 있는 상태이므로 삽입하지 않으며, 키의 중복을 허용하지 않음
- 삽입 연산도 반복 구조로도 구현할 수 있으나, 순환으로 구현하는 것이 가장 자연스러움

#### 삭제 연산
- 노드의 삭제 후에도 이진탐색트리의 특성을 반드시 유지해야하기 때문에 노드를 삭제하는 것은 이진탐색트리에서 가장 복잡한 연산
- 삭제연산은 삭제할 노드의 종류에 따라 3가지로 구분됨
 - case 1. 단말 노드의 삭제
 - case 2. 자식이 하나인 노드의 삭제
 - case 3. 두 개의 자식을 모두 갖는 노드의 삭제
- 단순연결리스트에서와 비슷하게 링크 표현법으로 나타낸 트리에서도 어떤 노드를 삭제하기 위해서는 부모 노드를 알아야 함
- 삭제할 노드의 자식이 하나뿐이면 자신 대신에 유일한 자식을 부모 노드에 연결해주면 됨
- 모든 경우에 대한 삭제연산 

#### 이진탐색트리의 성능 분석(p.335)
- 트리의 높이를 h라 하면 이진탐색트리의 탐색, 삽입, 삭제 연산의 시간 복잡도는 O(h)
- n개의 노드를 가지는 포화이진트리의 높이는 log2n이므로 최선의 경우 이진탐색트리 연산들의 시간복잡도는 O(log2n)
- 순차 탐색의 O(n)과 비교하면 탁월한 것
- 만약 한쪽으로 치우치는 완전 경사 이진트리일 경우 트리의 높이 h가 n과 같아지고, 탐색, 삭제, 삽입시간이 선형 탐색과 같은 O(n)이 됨
- 이진탐색트리의 효율을 높이기 위해서는 가능한 한 트리가 좌우로 균형 잡히게 만들어야 함

```python
# 이진탐색트리를 위한 노드 클래스
class BSTNode:            
    def __init__ (self, key, value):
        self.key = key
        self.value = value
        self.left = None
        self.right = None
```

```python
# 이진탐색트리 탐색연산(순환함수)
def search_bst(n, key) :
    if n == None :
        return None
    elif key == n.key:
        return n
    elif key < n.key:        
        return search_bst(n.left, key)
    else:  
        return search_bst(n.right, key)
```

```python
# 이진탐색트리 탐색연산(반복함수)
def search_bst_iter(n, key) :
    while n != None :        
        if key == n.key:        
            return n
        elif key < n.key:        
            n = n.left        
        else:            
            n = n.right       
    return None 
```

```python
# 이진탐색트리 탐색연산(preorder 사용): 값을 이용한 탐색
def search_value_bst(n, value) :
    if n == None : return None
    elif value == n.value:
        return n
    res = search_value_bst(n.left, value) 
    if res is not None :
       return res
    else :
       return search_value_bst(n.right, value)
```

```python
# 최대 값, 최소 값의 노드 탐색
def search_max_bst(n) :
    while n != None and n.right != None:
        n = n.right
    return n

def search_min_bst(n) :
    while n != None and n.left != None:
        n = n.left
    return n
```

```python
# 이진탐색트리 삽입연산(노드를 삽입함): 순환구조 이용
def insert_bst(r, n) :
    if n.key < r.key:
        if r.left is None :
           r.left = n
           return True
        else :
           return insert_bst(r.left, n)
    elif n.key > r.key :
        if r.right is None :
           r.right = n
           return True
        else :
           return insert_bst(r.right, n)
    else :         
        return False
```

```python
def delete_bst_case1 (parent, node, root) :
    if parent is None:     
        root = None        
    else :
        if parent.left == node :
            parent.left = None
        else :    
            parent.right = None

    return root     

def delete_bst_case2 (parent, node, root) :
    if node.left is not None :
        child = node.left
    else :
        child = node.right

    if node == root :
        root = child
    else :
        if node is parent.left : 
            parent.left = child
        else :
            parent.right = child

    return root	

def delete_bst_case3 (parent, node, root) :
    succp = node
    succ = node.right
    while (succ.left != None) :
        succp = succ
        succ = succ.left

    if (succp.left == succ) :
        succp.left = succ.right
    else :
        succp.right = succ.right

    node.key = succ.key
    node.value= succ.value
    node = succ;        

    return root           
```

```python
# 이진탐색트리 삭제연산(노드를 삭제함)
def delete_bst (root, key) :
    if root == None : return None       

    parent = None                       
    node = root                             
    while node != None and node.key != key :
        parent = node
        if key < node.key : node = node.left
        else : node = node.right;

    if node == None : return None       
    if node.left == None and node.right == None:
        root = delete_bst_case1 (parent, node, root)
    elif node.left==None or node.right==None :
        root = delete_bst_case2 (parent, node, root)
    else :
        root = delete_bst_case3 (parent, node, root)
```

### 심화 학습: 균형이진탐색트리
- 이진탐색트리에서는 균형을 유지하는 것이 무엇보다 중요

#### AVL 트리란?
- :모든 노드에서 왼쪽 서브트리와 오른쪽 서브트리의 높이 차가 1을 넘지 않는 이진탐색트리(모든 노드의 균형 인수는 0이나 +-1이 되어야 함)
 - 항상 균형 트리를 보장하기 때문에 탐색과 삽입, 삭제 연산에서 항상 O(logn)의 처리시간을 보장함
 - AVL 트리를 정이하기 위해서는 먼저 균형 인수를 정의해야 함
 - 어떤 노드의 균형 인수를 왼쪽 서브트리의 높이와 오른쪽 서브트리의 높이 차로 정의됨
 - AVL 트리는 비균형 상태가 되면 스스로 노드들을 재배치하여 균형 사태로 만듦
- AVL 트리에서의 탐색은 이진탐색트리의 탐색 연산과 동일함
 - 균형을 깨질 수 있는 연산은 삽입과 삭제 연산
- AVL 트리의 삽입 연산
 - 노드가 삽입되면 삽입되는 위치에서 루트까지의 경로에 있는 모든 조상 노드들의 균형 인수가 영향을 받음
 - 불균형 상태(균형 인수가 +-2)로 변한 가장 가까운 조상 노드의 서브 트리들에 대해 다시 균형을 잡하야 하며 그 외의 다른 노드들을 변경할 필요가 없음
- 삽입 연산에서 균형이 깨지는 경우는 4가지가 있음(새로 삽입된 노드 N으로부터 가장 가까우면서 균형 인수가 +-2가 된 조상 노드를 A라 하면)
 - LL 타입: N이 A의 왼쪽 자식의 왼쪽 서브 트리에 삽입된다
 - LR 타입: N이 A의 왼쪽 자식의 오른쪽 서브 트리에 삽입된다
 - RR 타입: N이 A의 오른쪽 자식의 오른쪽 서브 트리에 삽입된다
 - RL 타입: N이 A의 오른쪽 자식의 왼쪽 서브 틜에 삽입된다
- 각 방법에 대해 다시 균형을 잡는 회전 방법이 존재
 - LL과 RR 회전은 한번만 회전시키는 것을 단순 회전 방법(탐색 순서를 유지하면서 부모와 자식의 위치를 교환)
 - LR과 RL은 두 번의 회전이 필요한 이중 회전
- 재균형 함수
 - 노드의 균형을 검사하여 균형이 깨져 있으며 재균형하는 전체 함수는 노드가 삽입된 위치에서부터 루트까지의 경로상의 노드들에 대해 순서적으로 적용되어야 함
 - 반드시 서브트리의 새로운 루트를 반환해야함
- 삽입 함수 
 - 삽입 함수는 이진탐색트리의 삽입함수와 유사하지만 두 가지 차이가 존재
 - 삽입이 끝나면 반드시 재균형 함수를 호출해서 균형을 확인해야 함
 - 이진탐색트리에서와 달리 삽입 후 서브트리의 루트노드가 변경되는 경우가 많으므로 루트를 반환해야 함

#### AVL 트리를 이용한 맵

```python
def rotateLL(A) :
    B = A.left  
    A.left = B.right
    B.right = A
    return B

def rotateRR(A) :
    B = A.right    
    A.right = B.left
    B.left = A
    return B

def rotateRL(A) :
    B = A.right
    A.right = rotateLL(B)
    return rotateRR(A)

def rotateLR(A) :
    B = A.left
    A.left = rotateRR(B)
    return rotateLL(A)
```

```python
def calc_height_diff(n) :
    if n==None :
       return 0
    return calc_height(n.left) - calc_height(n.right)
```

```python
def reBalance (parent) :
    hDiff = calc_height_diff(parent)

    if hDiff > 1 :
        if calc_height_diff( parent.left ) > 0 :
            parent = rotateLL( parent )
        else :
            parent = rotateLR( parent )
    elif hDiff < -1 :
        if calc_height_diff( parent.right ) < 0 :
            parent = rotateRR( parent )
        else :
            parent = rotateRL( parent )
    return parent
```

```python
def insert_avl(parent, node) :
    if node.key < parent.key :
        if parent.left != None :
            parent.left = insert_avl(parent.left, node)
        else :
            parent.left = node
        return reBalance(parent)

    elif node.key > parent.key :
        if parent.right != None :
            parent.right = insert_avl(parent.right, node)
        else :
            parent.right = node
        return reBalance(parent)
    else :
        print("중복된 키 에러")
```

```python
class AVLMap(BSTMap):
    def __init__ (self):
        super().__init__()

    def insert(self, key, value=None):
        n = BSTNode(key, value)
        if self.isEmpty() :
           self.root = n
        else :
           self.root = insert_avl(self.root, n)

    def display(self, msg = 'AVLMap :'):
        print(msg, end='')
        levelorder(self.root)
        print()



node = [7,8,9,2,1,5,3,6,4]
map = AVLMap()

for i in node :
    map.insert(i)
    map.display("AVL(%d): "%i)

print(" 노드의 개수 = %d" % count_node( map.root ))
print(" 단말의 개수 = %d" % count_leaf( map.root ))
print(" 트리의 높이 = %d" % calc_height( map.root ))
```
